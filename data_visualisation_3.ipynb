```
import pandas as pd  # importing pandas library
import matplotlib.pyplot as plt  #importing pyplot from matplolib
import numpy as np   # importing numpy library
```

```
df=pd.read_csv("http://13.234.66.67/summer19/datasets/bank.csv")  # data read by using pandas library
df.head(10)
```

```
df.shape
```

```
# pie chart showing % of persons from different countries
pie_geo=df.groupby('Geography').NumOfProducts.count()   
rslt=plt.pie(pie_geo,labels=('france','spain','Germany'),autopct='%1.1f%%')
```

```
# density curve of credit score
df.head(50).CreditScore.plot(kind='kde')
```

```
# horizontal bar graph b/w age and geography
total=df.groupby('Age').NumOfProducts.count()

df.head(10).plot.barh('Geography','Age')
```

```
# line graph of no.of employes v/s country 
total=df.groupby('Geography').NumOfProducts.count()  
countries=('france', 'germany', 'spain')
plt.plot(countries,total)
plt.grid()
```

```
#piechart (in %) of males and females working
total=df.groupby('Gender').NumOfProducts.count()  
result=plt.pie(total,labels=('female', 'male'), explode=(0.05,0.05),shadow=True,autopct='%1.1f%%')
```

```
 # density curve of balance
df.head(50).Balance.plot(kind='kde')  
```

```
#density curve of salary
df.head(50).EstimatedSalary.plot(kind='kde') 
```

```
# age and salary bar graph showing highest salary paid to a specific age
import matplotlib.ticker as ticker
est=df.head(100).EstimatedSalary   
ag=df.head(100).Age

plt.bar(ag,est,width=0.8)
plt.xlim(18,50)
z=plt.subplot().xaxis.set_ticks(np.arange(15,65,5))
```

```
#age and estimated salary  scatter graph
est=df.head(100).EstimatedSalary  
ag=df.head(100).Age

plt.scatter(ag,est)

plt.grid()
plt.xlim(20,50)
plt.ylim(0,250000)
```

```
# scatter graph b/w tenure and balance of first 100 entries
bal=df.head(100).Balance  
ten=df.head(100).Tenure
plt.scatter(ten,bal)

plt.grid()
#plt.xlim(0,11)
#plt.ylim(0,200000)
```
