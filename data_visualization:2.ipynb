```python
import time
import pyspeedtest
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'  #rc--> run and configure(.bashrc)
#rcParams used for handling default matplolib values
```

```python
st = pyspeedtest.SpeedTest()
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()

i = 0
x, y = [], []   #list made to store time(in sec) in x and net speed in y

while True:
    x.append(i)   #store time in list x
    y.append(st.download())   #stores net speed at every second in list y 
    
    ax.plot(x, y, color='b')  #plots graph b/w x and y and color gives output of blue color line
    
    fig.canvas.draw() # draw() is used to make interactive model, it allows graph itself to change
    
    ax.set_xlim(left=max(0, i-50), right=i+50)   # to make the graph continuos otherwise graph will shrink
    #ax.set_ylim(0,25)
    time.sleep(0.1)  # time delay of 1 sec to show next plot on graph
    i += 1  
```

```python
plt.close()
```
