```
import requests
from bs4 import BeautifulSoup
import collections
import pandas as pd
import seaborn as sns
```

```
url='https://en.wikipedia.org/wiki/Machine_learning'
response=requests.get(url)
```

```
text=response.text
```

```
soup= BeautifulSoup(text, 'html')
```

```
soup
```

```
a=soup.find_all('p')
list1=[]
for i in a:
          z=i.text.split()
          for j in z:
            list1.append(j)
```

```
len(list1)
```

```
print(list1)
```

```
word_counter = collections.Counter(list1)
for word, count in word_counter.most_common():    #here if the arguement is 20 it decides the top most common words to be searched, if blank it searches all the words
    print(word, ": ", count)

print(word_counter.most_common(20))   #this prints the top 20 most occured words with their frequency
```

```
allwords = word_counter.most_common()
df = pd.DataFrame(allwords, columns = ['Word', 'Count'])
df.shape
```

```
df.head(20).Count.groupby(df.Word).sum().plot(kind='pie', radius=2.5, autopct='%1.1f%%')
```

```
df.head(20).plot.bar(x='Word',y='Count', figsize=(15,5))
```

```
#now filter data having count>3----pending

z=sns.lmplot('Word', 'Count', data=df.head(50), fit_reg=False, aspect=40/10  ,)

z.set_xticklabels(rotation=90, ha="right")
```
