```
x=[2,3,6]
y=[9,5,3]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)  
plt.grid()  #to see grid on graph
```

```
x=[2,3,6]
y=[9,5,3]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)  
plt.grid(color="cyan")  #to see grid on graph
```

```
x=[2,3,6]
y=[9,5,3]
x1=[2,4,1]
y1=[5,3,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)
plt.plot(x1,y1)
plt.grid(color="cyan")  #to see grid on graph
```

```
x=[2,3,6]
y=[9,5,3]
x1=[2,4,1]
y1=[5,3,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,label="water")  # draw a straight line
plt.plot(x1,y1,label="sand")  
plt.legend()  # to show label withn plot
plt.grid(color="cyan")  #to see grid on graph
```

```
x=[2,3,6]
y=[9,5,3]
x1=[2,4,1]
y1=[5,3,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,label="water")  # draw a straight line
plt.plot(x1,y1,label="sand")  
plt.legend()            # to show label withn plot
plt.grid(color="cyan")        #to see grid on graph
plt.xlim(0,10)
plt.ylim(0,10)           # to show min and max in y axis
```

```
x=[2,3,6]
y=[9,5,3]
x1=[2,4,1]
y1=[5,3,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,label="water")  # draw a straight line
plt.plot(x1,y1,label="sand") 
plt.bar(x1,y1)
plt.bar(x,y)
plt.legend()            # to show label withn plot
plt.grid(color="cyan")        #to see grid on graph
plt.xlim(0,10)
plt.ylim(0,10)           # to show min and max in y axis
```
