```
newdata1=[i for i in mydata.split()]
```

```

#NLTK graphs
import matplotlib.pyplot as plt
import nltk
%matplotlib inline # incase matplotlib is not supporting by default with pandas or nltk 
#to replace plt.show() from any platform
```

```
[i for i in dir(nltk) if 'Freq' in i]
```

```
nlpdata=nltk.FreqDist(newdata1)
```

```
nlpdata.plot(20)   #top 20 words
```

```
#we can remove stopwords from newdata1 then plot graph from nltk.corpus import stopwords

#removing stopwords
removedata=[i for i in newdata1 if i.lower() not in stopwords.words('english')]    #in stopwords all keywords are in small .lower() is to make all letters(Is, Then,etc) in small letters
```

```
nlpremove=nltk.FreqDist(removedata)
```

```
nlpremove.plot(20)
```
